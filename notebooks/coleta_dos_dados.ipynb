# Tratamento das tabelas

## copiar para DFs singulares

primeiramente vamos passar todos para DFs individuais para melhor manuseio

```python
df_daumau = dfs['daumau']
df_desinstalacoes = dfs['desinstalacoes']
df_installs = dfs['installs']
df_ratings_reviews = dfs['ratings_reviews']
```

## converter para datetime

garantindo que todos tem o mesmo formato de data

```python
#conferindo se há date acima de hoje em outros
hoje = pd.to_datetime('2025-06-11')
print('em desinstalações', len(df_desinstalacoes[df_desinstalacoes['date'] > hoje]))
print('em installs', len(df_installs[df_installs['date'] > hoje]))
print('em ratings_reviews', len(df_ratings_reviews[df_ratings_reviews['date'] > hoje]))
```

```python
#converter datas de df_daumau e df_ratings_reviews em datetime64[ns]
df_daumau['date'] = pd.to_datetime(df_daumau['date'])
df_ratings_reviews['date'] = pd.to_datetime(df_ratings_reviews['date'])

#filtrar dados depois de 2025-06-11
date_threshold = pd.to_datetime('2025-06-11')
df_daumau = df_daumau[df_daumau['date'] <= date_threshold]

#conferir
print(dfs['daumau'].info())
print(dfs['ratings_reviews'].info())
```

## remover colunas

`mauReal` não é a target, então ela pode atrapalhar o modelo pela semelhança. E `country` e `lang` são de apenas um valor sem variabilidade

```python
#remover colunas country e lang de df_desinstalacoes
df_desinstalacoes = df_desinstalacoes.drop(columns=['country', 'lang'])

#remover mauReal de df_daumau
df_daumau = df_daumau.drop(columns=['mauReal'])

#conferindo
print(df_desinstalacoes.columns)
print(df_daumau.columns)
```

## remover duplicados

```python
#renomeando colunas de 'appid' para 'appId'
df_installs = df_installs.rename(columns={'appid': 'appId'})
df_ratings_reviews = df_ratings_reviews.rename(columns={'appid': 'appId'})

# remover duplicados em appId e date de todos
df_daumau = df_daumau.drop_duplicates(subset=['appId', 'date'])
df_desinstalacoes = df_desinstalacoes.drop_duplicates(subset=['appId', 'date'])
df_installs = df_installs.drop_duplicates(subset=['appId', 'date'])
df_ratings_reviews = df_ratings_reviews.drop_duplicates(subset=['appId', 'date'])

#resetar o index de todos
df_daumau = df_daumau.reset_index(drop=True)
df_desinstalacoes = df_desinstalacoes.reset_index(drop=True)
df_installs = df_installs.reset_index(drop=True)
df_ratings_reviews = df_ratings_reviews.reset_index(drop=True)

#checando se tem duplicado em appId e date
print(df_daumau.duplicated(subset=['appId', 'date']).sum())
print(df_desinstalacoes.duplicated(subset=['appId', 'date']).sum())
print(df_installs.duplicated(subset=['appId', 'date']).sum())
print(df_ratings_reviews.duplicated(subset=['appId', 'date']).sum())
```

## analisando faltantes

```python
#info de todos os nulos
print("Nulos em daumau:")
print(df_daumau.isnull().sum(), '\n')
print("Nulos em desinstalacoes:")
print(df_desinstalacoes.isnull().sum(), '\n')
print("Nulos em installs:")
print(df_installs.isnull().sum(), '\n')
print("Nulos em ratings_reviews:")
print(df_ratings_reviews.isnull().sum())
```

como apenas `dauReal` apresentou nulos, e ela é nossa target, iremos remover os nulos

```python
#remover nulos em daumau
df_daumau = df_daumau.dropna()

#resetar_index
df_daumau = df_daumau.reset_index(drop=True)

#conferindo
df_daumau.isnull().sum()
```

# merge

iremos agora unir as tabelas com intuito de agregar informações à tabela com a target, pois nos próximos passos desse projeto, será avaliado quais métricas e variáveis influenciam na target, o que será relevante para o modelo.

## fazendo o merge

```python
merged_df = df_daumau.merge(df_desinstalacoes, on=['appId', 'date'], how='left')
merged_df = merged_df.merge(df_installs, on=['appId', 'date'], how='left')
merged_df = merged_df.merge(df_ratings_reviews, on=['appId', 'date'], how='left')

#ordenar por data
merged_df = merged_df.sort_values(by='date')

#resetar o index
merged_df = merged_df.reset_index(drop=True)

print("Shape:", merged_df.shape)
merged_df.head()
```

```python
#remover todos com data anterior a 2008-07-10 (criação da app store)
merged_df = merged_df[merged_df['date'] >= '2008-07-10']

#resetar o index
merged_df = merged_df.reset_index(drop=True)

print("Shape:", merged_df.shape)
merged_df.head()
```

## tratando os nulos

```python
#contabilizando nulos
merged_df.isnull().sum()
```

como não há nulos na target. E as com faltantes pode siginificar que não houve a ocorrência para tal app tal dia ou a taxa estava zero, iremos zerar todos os nulos, exceto a variável categórica que será descrita com uma informação equivalente como `DESCONHECIDO`

```python
#preencher nulos em category como 'DESCONHECIDO'
merged_df['category'] = merged_df['category'].fillna('DESCONHECIDO')

#preencher todoas as numéricas com 0
numericas = ['predictionLoss', 'newinstalls', 'ratings', 'daily_ratings', 'reviews', 'daily_reviews']
merged_df[numericas] = merged_df[numericas].fillna(0)

#conferindo
print(merged_df.isnull().sum())
print(merged_df.info())
```
